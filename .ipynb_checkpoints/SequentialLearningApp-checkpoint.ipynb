```python
#EventHandling

def content_parser():
    if up.value == {}:
        with out:
            display(Markdown('No CSV loaded'))
            #print('No CSV loaded')    
    else:
        typ, content = "", ""
        up_value = up.value
        for i in up_value.keys():
            typ = up_value[i]["metadata"]["type"]
            if typ == "text/csv":
                content = up_value[i]["content"]
                content_str = str(content, 'utf-8')

                if eraser.value != {}: 
                    for val in eraser.value:
                        if val == "tab":
                            content_str = content_str.replace("\t","")
                        else:
                            content_str = content_str.replace(val,"")
                if content_str != "":
                    str_io = StringIO(content_str) 
                    return str_io
def df_converter():
    content = content_parser()
    if content is not None:
            df = pd.read_csv(content, sep=delim.value, index_col=False, skiprows=rows.value)
            
            
                
            return df
    else:
        return None
def preview():
    
    df = df_converter()
    with out:
        out.clear_output()
        display(Markdown('Now this is how your DF looks like:'))
        
        if df is not None:
            print(df.head(10))
        else:
            display(Markdown('Configuration is wrong/missing...'))
            
def upload():
    
    df = df_converter()
    with out:
        out.clear_output()
        display(Markdown('Your uploaded DF looks like:'))
       
        if df is not None:
            print(df)
            x_axis.options = df.columns
            y_axis.options = df.columns
            feature_selector.options= df.columns
            select_x.options=df.columns
            select_y.options=df.columns
            select_size.options=df.columns
            select_hue.options=df.columns
            
        else:
            display(Markdown('Configuration is wrong/missing...'))
            #print('Configuration is wrong/missing...')
def desc():
    info_level = toggle.value
    if info_level != {}:
        df = df_converter()
        with out:
            out.clear_output()
            print('\n ------Your {} looks like:------ \n'.format(
                info_level))
            if df is not None:
                if info_level == 'Info  ':
                    print(df.info(verbose=True))
                elif info_level == 'Stats  ':
                    print(df.describe())
                elif info_level == 'Preview  ':
                    print(df.head(5))
                else:
                    display(Markdown('Configuration is wrong/missing...'))
    
    
def simplify_colomns():
    content = content_parser()
    if content is not None:
            df = pd.read_csv(content, sep=delim.value, index_col=False, skiprows=rows.value)
            
            df.columns=df.columns.str.replace(r"\(.*\)","")
            df.columns=df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
                
           
            with out:
                out.clear_output()
                display(Markdown('Now this is how your DF looks like:'))
                if df is not None:
                    print(df.head(10))
                else:
                    display(Markdown('Configuration is wrong/missing...'))
                return df
    else:
        return None
    
    
#plot methods   

def plot():
    graph = graph_type.value
    if graph != {}:
        df = df_converter()
        with out:
            out.clear_output()
            print('\n ------Your {} looks like:------ \n'.format(
                graph))
            if (df is not None):
                df = df.head(5)
                height = df[y_axis.value]
                bars = df[x_axis.value]
                y_pos = np.arange(len(height))
                plt.figure(figsize=(10,4))
                if graph == 'Bar Chart':
                    plt.bar(
                        y_pos, 
                        height, 
                        color=color_picker.value)
                    plt.xticks(y_pos, bars)
                elif graph == 'Line Chart':
                    plt.plot(
                        bars,
                        height, 
                        color=color_picker.value,
                        marker='o', 
                        linestyle='solid'
                    )
                    plt.xticks(bars)
                plt.show()

                
def plot_pairwise():
    
    df = df_converter()
    with out_pairwise:
        sns.pairplot(df)
        plt.show()

def plot_heat():
    df = df_converter()
    with out_heat:
        corr = df.corr()

        plt.figure(figsize=(9,7))
        sns.heatmap(corr, annot=True, cmap='Blues')
        b, t = plt.ylim()
        plt.ylim(b+0.5, t-0.5)
        plt.title("Feature Correlation Heatmap")
        plt.show()
            
def plot_scatter():
    data=df_converter()
    with out_scatter:
        fig, ax = plt.subplots(figsize=(10,7))
        #not generic
        sns.scatterplot(y=select_y.value, x=select_x.value, hue=select_hue.value, size=select_size.value, data=data, ax=ax, sizes=(50, 300))
        ax.set_title(select_y.value+ "vs"+ select_x.value)
        ax.legend(loc="upper left", bbox_to_anchor=(1,1))
        plt.show()
    
```

```python
#ml eventhandling
#pre pocessing


def create_dynamically_checkboxes():
    target= target_selection.value
    items = [Checkbox(description=feature,value=True,disabled=False) for feature in target]
    box_features.children=items
    

def create_dynamically_sliders():
    target= target_selection.value
    df = df_converter()
    
    Y = df.loc[:,df.columns.isin(feature_selector.value)]
    
    items = [FloatRangeSlider(
    value=[0,100],
    min=0,
    max=100,
    step=5,
    description='Quantile:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',) for feature in target ]
    
    box_features_slider.children=items
    
    
def create_dynamically_button():
    box_confirm.children=[button_confirm_options]

def confirm_options():
    items=box_features_slider.children
    df = df_converter()
    Y = df.loc[:,df.columns.isin(target_selection.value)]
    
    for slider in items:
            unt_grenz= slider.value[0]/100
            ob_grenz= slider.value[1]/100
            Y = Y[(Y >= Y.quantile(unt_grenz) ) & (Y <= Y.quantile(ob_grenz))]
            Y= Y.dropna()
    
    return Y
    
def confirm_features():
    df = df_converter()
    train = feature_selector.value
    
    target_selection.options=df.columns[~df.columns.isin(feature_selector.value)]
    
    with out_data_pre_pr:
        out_data_pre_pr.clear_output
        display(Markdown('**Your selected Features are:**'))
        for feature in train:
              display(Markdown(feature))
       
    

def get_features():
    df = df_converter()
    train = feature_selector.value
    
    target_selection.options=df.columns[~df.columns.isin(feature_selector.value)]
    return train
    
def confirm_target():
    target_Var=target_selection.value
    #feature_selector.options= get_features()
    if len(target_Var)<len(get_features()):
        create_dynamically_checkboxes()
        create_dynamically_sliders()
        create_dynamically_button()

    with out_data_pre_pr:
        out_data_pre_pr.clear_output
        if len(target_Var)<len(get_features()):
            display(Markdown('**Your chosen Target Variable is/are:**'))
            for feature in target_Var:
                  display(Markdown(feature))
        else:
            display(Markdown('**ItÂ´s not possible that the number of targets is greater than the number of training Features**'))
            display(Markdown('**Please reselect!**'))
        
    
    
def split_data(self):
    #debug
    
    df=df_converter()
    X=df.loc[:,df.columns.isin(feature_selector.value)]
    Y= confirm_options()
    
    from sklearn.model_selection import train_test_split

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
    
    
    with out_data_pre_pr:
            out_data_pre_pr.clear_output
            display(Markdown('You splittet your Target-Sample and Feature-Sample into training and test sets:'))
           
    sets=[X_train, X_test, Y_train, Y_test]
    
    return sets

def scale_data(self):
    sets= split_data(self)
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    X_train= sets[0]
    X_test=sets[1]
    
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    with out_data_pre_pr:
        display(Markdown('Scaling Finished'))
        

        
```

```python
#Algos
class Algos:
    
    def __init__(self, dataframe, lr_alg, lasso_alg, ridge_alg,des_tree_alg,rand_for_alg):  #constructor
        self.dataframe = dataframe
        self.lr_alg=lr_alg
        self.lasso_alg=lasso_alg
        self.ridge_alg=ridge_alg
        self.des_tree_alg=des_tree_alg
        self.rand_for_alg=rand_for_alg
   
    def compare(self):
        sets=split_data(self)
        X_train = sets[0]
        X_test=sets[1]
        y_train = sets[2]
        y_test = sets[3]
        
        models = [lr, lasso, ridge, dtr, rfr]
        names = ["Linear Regression", "Lasso Regression", "Ridge Regression", 
                 "Decision Tree Regressor", "Random Forest Regressor"]
        rmses = []
        with out_compare_plot:
            out_compare_plot.clear_output()
            for model in models:
                rmses.append(np.sqrt(mean_squared_error(y_test, model.predict(X_test))))

            x = np.arange(len(names)) 
            width = 0.3

            fig, ax = plt.subplots(figsize=(10,7))
            rects = ax.bar(x, rmses, width)
            ax.set_ylabel('RMSE')
            ax.set_xlabel('Models')
            ax.set_title('RMSE with Different Algorithms')
            ax.set_xticks(x)
            ax.set_xticklabels(names, rotation=45)
            self.autolabel(rects,ax)
            fig.tight_layout()
            plt.show()
        
    #sets=[X_train, X_test, Y_train, Y_test]
    def des_tree(self):
        sets=split_data(self)
        X_train = sets[0]
        X_test=sets[1]
        y_train = sets[2]
        y_test = sets[3]
        #dtr = DecisionTreeRegressor()
        dtr.fit(X_train, y_train)
        
        y_pred_dtr = dtr.predict(X_test)
        with out_algo_alt:
            out_algo_alt.clear_output
            print("Model\t\t\t\t RMSE \t\t MSE \t\t MAE \t\t R2")
            print("""Decision Tree Regressor \t {:.2f} \t\t {:.2f} \t\t{:.2f} \t\t{:.2f}""".format(
                        np.sqrt(mean_squared_error(y_test, y_pred_dtr)),mean_squared_error(y_test, y_pred_dtr),
                        mean_absolute_error(y_test, y_pred_dtr), r2_score(y_test, y_pred_dtr)))

            plt.scatter(y_test, y_pred_dtr)
            plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
            plt.xlabel("Predicted")
            plt.ylabel("True")
            plt.title("Decision Tree Regressor")
            plt.show()
        
    
    def rand_for(self):
        sets=split_data(self)
        X_test=sets[1]
        X_train = sets[0]
        y_train = sets[2]
        y_test = sets[3]
        
        rfr.fit(X_train, y_train)

        y_pred_rfr = rfr.predict(X_test)
        with out_algo_alt:
            print("Model\t\t\t\t RMSE \t\t MSE \t\t MAE \t\t R2")
            print("""Random Forest Regressor \t {:.2f} \t\t {:.2f} \t\t{:.2f} \t\t{:.2f}""".format(
                        np.sqrt(mean_squared_error(y_test, y_pred_rfr)),mean_squared_error(y_test, y_pred_rfr),
                        mean_absolute_error(y_test, y_pred_rfr), r2_score(y_test, y_pred_rfr)))

            plt.scatter(y_test, y_pred_rfr)
            plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
            plt.xlabel("Predicted")
            plt.ylabel("True")
            plt.title("Random Forest Regressor")
            plt.show()
        
    
    def linreg(self):
        sets=split_data(self)
        
        # Linear Regression
       #lr = LinearRegression()
        X_train = sets[0]
        Y_train = sets[2]
        lr.fit(X_train, Y_train)
        Y_pred_lr = lr.predict(sets[1])
        
        with out_ml:
            print("Model\t\t\t RMSE \t\t MSE \t\t MAE \t\t R2")
            print("""LinearRegression \t {:.2f} \t\t {:.2f} \t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(sets[3], Y_pred_lr)),mean_squared_error(sets[3], Y_pred_lr),
            mean_absolute_error(sets[3], Y_pred_lr), r2_score(sets[3], Y_pred_lr)))
            
        return Y_pred_lr

    def lassoreg(self):
        sets=split_data(self)
        from sklearn.linear_model import Lasso

        X_train = sets[0]
        Y_train = sets[2]
        #lasso = Lasso()
        lasso.fit(X_train, Y_train)
        Y_pred_lasso = lasso.predict(sets[1])
        with out_ml:
                #could be refactored
            
            print("""Lasso \t {:.2f} \t\t {:.2f} \t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(sets[3], Y_pred_lasso)),mean_squared_error(sets[3], Y_pred_lasso),
            mean_absolute_error(sets[3], Y_pred_lasso), r2_score(sets[3], Y_pred_lasso)))
        return Y_pred_lasso

    def ridgereg(self):
        sets=split_data(self)
        from sklearn.linear_model import Ridge
        #ridge = Ridge()
        X_train = sets[0]
        Y_train = sets[2]
        ridge.fit(X_train, Y_train)
        Y_pred_ridge = ridge.predict(sets[1])
        with out_ml:
                #could be refactored
            
            print("""RidgeRegression \t {:.2f} \t\t {:.2f} \t{:.2f} \t\t{:.2f}""".format(
            np.sqrt(mean_squared_error(sets[3], Y_pred_ridge)),mean_squared_error(sets[3], Y_pred_ridge),
            mean_absolute_error(sets[3], Y_pred_ridge), r2_score(sets[3], Y_pred_ridge)))
        return Y_pred_ridge
##plot
   
    def plot_coef(self):
        df=df_converter()
        coeff_lr = lr.coef_.squeeze()
        coeff_lasso = lasso.coef_.squeeze()
        coeff_ridge = ridge.coef_.squeeze()
        with out_algo:
            labels = df.columns[:-1].tolist()
            
            x = np.arange(len(labels))
            width = 0.3
            x_graph=x - 2*(width/2)
            fig, ax = plt.subplots(figsize=(10,10))
            rects1 = ax.bar(x_graph, coeff_lr, width, label='LR')
            rects2 = ax.bar(x, coeff_lasso, width, label='Lasso')
            rects3 = ax.bar(x + 2*(width/2), coeff_ridge, width, label='Ridge')

            ax.set_ylabel('Coefficient')
            ax.set_xlabel('Features')
            ax.set_title('Feature Coefficients')
            ax.set_xticks(x)
            ax.set_xticklabels(labels, rotation=45)
            ax.legend()


            self.autolabel(rects1,ax)
            self.autolabel(rects2,ax)
            self.autolabel(rects3,ax)

            fig.tight_layout()
            plt.show()
            
    def autolabel(self,rects,ax):
            """Attach a text label above each bar in *rects*, displaying its height."""
            for rect in rects:
                height = rect.get_height()
                
                ax.annotate('{:.2f}'.format(height), xy=(rect.get_x() + rect.get_width() / 2, height),
                            xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')
    
    
```

```python
class sequential_learning:
    sets =[]
    rfpred=[]
    number_of_executions=0
    
    
    def __init__(self, dataframe, rf):  #constructor
        self.dataframe = dataframe
        self.rf=rf
    
    def main(self):
        split_data(self);
        number_of_executions=iterations.value
        strategy=confirm_strategy()
        if strategy == 'MEI':
            MEI(rfpred[0])
            
        if strategy == 'MU':
            MU(rfpred[1])
            
        if strategy == 'MLI':
            MLI(rfpred[0], rfpred[1])
        
            
            
    def split_data(self):
        df=dataframe
        X=df.loc[:,df.columns.isin(feature_selector.value)]
        y=target_selection.value
        index=df.columns.get_loc(target_selection.value)

        Y=df.iloc[:,index].to_frame()

        from sklearn.model_selection import train_test_split

        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)


        with out_data_pre_pr:
                out_data_pre_pr.clear_output
                display(Markdown('You splittet your Target-Sample and Feature-Sample into training and test sets:'))
        sets=[X_train, X_test, Y_train, Y_test]
        return sets
    
    
    
    # methods for selecting the next best candidate "to make an experiment with"  
    def predictions(X_train, y_train, X_test):
        """predict expected target value and the corresponding uncertainty of all X_test points"""
        #rf = RandomForestRegressor()
        rf.fit(X_train, y_train)
        y_pred, y_std = rf.predict(X_test, return_std=True)
        rfpred= [y_pred, y_std]
        return rfpred

    
    
    def MEI(y_pred):
        
        """outputs the X_test index with the biggest expected target value corresponding to it"""
        index_new_experiment = y_pred.argmax(axis=0)
        print("indexing mei")
        return index_new_experiment 

    def MU(y_std):
        """outputs the X_test index with the biggest uncertainty corresponding to it"""
        index_new_experiment = y_std.argmax(axis=0)
        print("indexing mu")
        return index_new_experiment

    def MLI(y_pred, y_std):
        """outputs the X_test value which gives the maximum likelyhood of improvement"""
        value = list( map(add, y_pred, y_std) )
        value = np.asarray(value)
        index_new_experiment = value.argmax(axis=0)
        print("indexing mli")
        return index_new_experiment
    
    
########Supportive Methods#############
    def building_new_training_and_test_sets(X_train, y_train, X_test, y_test, index, iteration, x_times_choosen, x_updated_index_array):
        """prepare the next training and testset for the next iteration"""
        #append to the X_train and y_train
        X_t = np.reshape(X_test[index], (1,8))
        X_train_new = np.append(X_train,X_t, axis= 0)
        y_train_new = np.append(y_train, y_test[index])
        #drop from X_test and y_test
        X_test_new = np.delete(X_test, obj = index, axis=0)
        y_test_new = np.delete(y_test, obj = index, axis=0)
        x_times_choosen[(x_updated_index_array[index])] = 1
        x_updated_index_array_new = np.delete(x_updated_index_array, obj = index)
        iteration = iteration +1
        return (X_train_new, y_train_new, X_test_new, y_test_new, iteration, x_times_choosen, x_updated_index_array_new)


    def finding_best_candidate(y, y_train, y_test, X_train, X_test, method, x_updated_index_array, stop_criterion_level = 0.8):
        """finding iteratively the candidate with the optimal target value"""
        iteration = 0 
        x_times_choosen = np.zeros(len(y))
        while y_train.max(axis=0) < stop_criterion_level*y.max(axis=0):
            y_pred, y_std = predictions(X_train, y_train, X_test)
            if method == MEI:
                index_new_experiment = MEI(y_pred)
            elif method == MU:
                index_new_experiment = MU(y_std)
            elif method == MLI:
                index_new_experiment = MLI(y_pred, y_std)
            X_train, y_train, X_test, y_test, iteration, x_times_choosen, x_updated_index_array= building_new_training_and_test_sets(X_train, y_train, X_test, y_test, index_new_experiment, iteration, x_times_choosen, x_updated_index_array)    
        return (iteration, X_train, y_train,X_test,y_test, x_times_choosen)

    def create_train_test_set(X,y,train_set_size=10, train_target_bound_level =0.5):
        x_updated_index_array_test = np.arange(len(y))
        y_train = []
        more_train_set_size = train_set_size
        X_test = X
        X_train_0 = []
        while  more_train_set_size != 0:
            X_train, X_test, x_updated_index_array_train, x_updated_index_array_test = train_test_split(X_test, x_updated_index_array_test, train_size=more_train_set_size, random_state=i)

            try:
                X_train_0 = np.concatenate((X_train_0, X_train))
            except:
                X_train_0 = X_train

            for updated_index, j in zip(np.flipud(x_updated_index_array_train), reversed(range(len(x_updated_index_array_train)))):
                for all_indices in range(len(y)):
                    if updated_index == all_indices:
                        if y[updated_index] < train_target_bound_level*y.max(axis=0):
                            y_train.append(y[updated_index])
                        else:
                            X_test = np.concatenate((X_test , np.reshape(X_train[j],(1,8))))
                            x_updated_index_array_test = np.concatenate((x_updated_index_array_test , np.reshape(x_updated_index_array_train[j],(1,))))
                            X_train_0 = np.delete(X_train_0, obj = j, axis=0)
                            x_updated_index_array_train = np.delete(x_updated_index_array_train, obj = j, axis=0)

            more_train_set_size = train_set_size - len(y_train)
            X = X_test
            x_updated_index_array = x_updated_index_array_test

        y_test = []
        for updated_index in x_updated_index_array_test:
            for all_indices in range(len(y)):
                if updated_index == all_indices:
                    y_test.append(y[all_indices])
        for updated_index in np.flipud(x_updated_index_array_train):
            for all_indices in range(len(y)):
                if updated_index == all_indices:
                    if y[updated_index] >= train_target_bound_level*y.max(axis=0):
                        y_test.append(y[updated_index])

        y_train = np.asarray(y_train)
        y_test = np.asarray(y_test)

        return X_train_0, X_test, y_train, y_test, x_updated_index_array_train, x_updated_index_array_test
#######Supportive Methods#############

            # Actual FUELS Script
            # Dimension reduction for latter visualization of results

        #tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
       # tsne_results = tsne.fit_transform(X)

            # MEI - Maximum expected improvement
            ###Togle
    def MEI_seq():
                number_of_executions = 30
                total_iterations_MEI = 0
                iterations_MEI = []
                x_times_choosen_MEI = np.zeros((len(y), 1))
                for i in range(number_of_executions):

                    X_train, X_test, y_train, y_test, x_updated_index_array_train, x_updated_index_array_test = create_train_test_set(X,y,train_set_size=10, train_target_bound_level =0.5)
                    iterations_needed, X_train, y_train, X_test, y_test, x_times_choosen = finding_best_candidate(y, y_train, y_test, X_train, X_test, MEI, x_updated_index_array_test)
                    for p in range(len(y)):
                        x_times_choosen_MEI[p] = x_times_choosen_MEI[p] + x_times_choosen[p]

                    total_iterations_MEI = total_iterations_MEI + iterations_needed
                    iterations_MEI.append(iterations_needed+10)
                iterations_mean_MEI = total_iterations_MEI/number_of_executions +10
                var_MEI = np.sum((np.array(iterations_MEI)-iterations_mean_MEI)**2)/(number_of_executions-1)
                std_MEI = np.sqrt(var_MEI)
                uncertainty_MEI = std_MEI/np.sqrt(number_of_executions)
                print(iterations_mean_MEI)
                print(uncertainty_MEI)

            # Plot Results in reduced FS

                plt.figure(figsize=(16,10))
                color = np.log([x_times_choosen_MEI+1])
                # color = [y]
                sc=plt.scatter(x=tsne_results[:,0], y=tsne_results[:,1], s=200, c=color)
                palette=sns.color_palette("hls")
                plt.colorbar(sc)


            # MLI - Maximum expected improvement
    def MLI_seq():
                number_of_executions = 30
                total_iterations_MLI = 0
                iterations_MLI = []
                x_times_choosen_MLI = np.zeros((len(y),1))
                for i in range(number_of_executions):
                    X,y = process_csv_datei("citrination-export (1).csv") 

                    X_train, X_test, y_train, y_test, x_updated_index_array_train, x_updated_index_array_test = create_train_test_set(X,y,train_set_size=10, train_target_bound_level =0.5)
                    iterations_needed, X_train, y_train,X_test, y_test, x_times_choosen = finding_best_candidate(y, y_train, y_test, X_train, X_test, MLI,x_updated_index_array_test)
                    for p in range(len(y)):
                        x_times_choosen_MLI[p] = x_times_choosen_MLI[p] + x_times_choosen[p]
                    #print(i)
                    #print(len(X_train), len(X_test), len(y_train), len(y_test))
                    total_iterations_MLI = total_iterations_MLI + iterations_needed
                    iterations_MLI.append(iterations_needed+10)
                iterations_mean_MLI = total_iterations_MLI/number_of_executions +10
                var_MLI = np.sum((np.array(iterations_MLI)-iterations_mean_MLI)**2)/(number_of_executions-1)
                std_MLI = np.sqrt(var_MLI)
                uncertainty_MLI = std_MLI/np.sqrt(number_of_executions)
                print(iterations_mean_MLI)
                print(uncertainty_MLI)

            # Plot Results in reduced FS

                plt.figure(figsize=(16,10))
                color = np.log([x_times_choosen_MLI+1])
                # color = [y]
                sc=plt.scatter(x=tsne_results[:,0], y=tsne_results[:,1], s=200, c=color)
                palette=sns.color_palette("hls")
                plt.colorbar(sc)
            # 


            # MU - Maximum Uncertainty
    def MU_seq():
            number_of_executions = 30
            total_iterations_MU = 0
            iterations_MU = []
            x_times_choosen_MU = np.zeros((len(y), 1))
            for i in range(number_of_executions):

                X_train, X_test, y_train, y_test, x_updated_index_array_train, x_updated_index_array_test = create_train_test_set(X,y,train_set_size=10, train_target_bound_level =0.5)
                iterations_needed, X_train, y_train,X_test, y_test, x_times_choosen = finding_best_candidate(y, y_train, y_test, X_train, X_test, MU,x_updated_index_array_test)
                for p in range(len(y)):
                    x_times_choosen_MU[p] = x_times_choosen_MU[p] + x_times_choosen[p]
                #print(i)
                #print(len(X_train), len(X_test), len(y_train), len(y_test), x_times_choosen_MU[141])
                total_iterations_MU = total_iterations_MU + iterations_needed
                iterations_MU.append(iterations_needed+10)
            iterations_mean_MU = total_iterations_MU/number_of_executions +10
            var_MU = np.sum((np.array(iterations_MU)-iterations_mean_MU)**2)/(number_of_executions-1)
            std_MU = np.sqrt(var_MU)
            uncertainty_MU = std_MU/np.sqrt(number_of_executions)
            print(iterations_mean_MU)
            print(uncertainty_MU)

            # Plot Results in reduced FS

            plt.figure(figsize=(16,10))
            color = np.log([x_times_choosen_MU+1])
            # color = [y]
            sc=plt.scatter(x=tsne_results[:,0], y=tsne_results[:,1], s=200, c=color)
            palette=sns.color_palette("hls")
            plt.colorbar(sc)


            ##############################################################################
            # Plot some Results: 


            x = np.arange(3)
            iterations = [iterations_mean_MLI,iterations_mean_MEI,iterations_mean_MU]
            uncertainties = [uncertainty_MLI,uncertainty_MEI,uncertainty_MU]
            couleur = ['#3776ab', '#5bc48b', '#dec948']

            fig, ax = plt.subplots()
            plt.bar(x, iterations, color= couleur, yerr= uncertainties)
            plt.xticks(x, ('MLI', 'MEI', 'MU'))
            plt.show()

            plt.figure(figsize=(10,7))
            color = [y]
            sc = plt.scatter(x=tsne_results[:,0], y=tsne_results[:,1], s=200, c=color)
            palette=sns.color_palette("hls")
            plt.colorbar(sc)



            vis_mat_mei=data.values[:,8]
            vis_mat_mei = np.column_stack((vis_mat_mei, x_times_choosen_MEI))
            vis_mat_mei = np.column_stack((vis_mat_mei, x_times_choosen_MLI))
            vis_mat_mei_sort = vis_mat_mei[np.argsort(vis_mat_mei[:, 0])]

            ax.set_ylabel('Importance')
            ax.set_xlabel('Features')
            ax.set_title('Feature Importance')
            ax.set_xticks(x)
            ax.set_xticklabels(labels, rotation=45)
            ax.legend(loc="upper left", bbox_to_anchor=(1,1))
```
